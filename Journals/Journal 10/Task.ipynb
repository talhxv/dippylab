# Using Horizontal Filter

```python

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('Image.png')
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
kernel = np.array([[-1,-1,-1],
                  [2,2,2],
                  [-1,-1,-1]])

img_blur = cv.GaussianBlur(img_gray, (3,3), 0)
output = cv.filter2D(src=img_blur, kernel=kernel, ddepth=-1 )
plt.imshow(cv.cvtColor(output, cv.COLOR_BGR2RGB))


```

# Using Manual Filters

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('Image2.png')
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
kernel = np.array([[-1,-1,-1], # This is horizontal my boy
                  [-1,2,-1],
                  [-1,-1, -1]])
kernel2 = np.array([[-1,2,-1], # This is Vertical
                  [-1,2,-1],
                  [-1,2, -1]])

img_blur = cv.GaussianBlur(img_gray, (3,3), 0)
outputx = cv.filter2D(src=img_blur, kernel=kernel2, ddepth=-1 )
outputy= cv.filter2D(src=img_blur, kernel=kernel, ddepth=-1)

output = cv.add(outputx,outputy)

plt.imshow(cv.cvtColor(output, cv.COLOR_GRAY2RGB))
```

# Using Sobel Filter

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('Image2.png')
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_blur = cv.GaussianBlur(img_gray, (3,3), 0)
sobelx = cv.Sobel(src=img_blur, ddepth=-1, dx=1, dy=0, ksize=5) 
sobely = cv.Sobel(src=img_blur, ddepth=-1, dx=0, dy=1, ksize=5) 
sobelxy = cv.Sobel(src=img_blur, ddepth=-1, dx=1, dy=1, ksize=5)

plt.imshow(cv.cvtColor(sobelxy, cv.COLOR_BGR2RGB))
```

# Task where manually analysing what to do 

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('Image3.png')
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_blur = cv.GaussianBlur(img_gray, (3,3), 0)

sobelxy = cv.Sobel(src=img_blur, ddepth=-1, dx=1, dy=1, ksize=5)

ekernel = np.ones((2, 2), np.uint8)
 
dilated = cv.dilate(sobelxy, ekernel, iterations=2)

plt.imshow(cv.cvtColor(dilated, cv.COLOR_BGR2RGB))
```
