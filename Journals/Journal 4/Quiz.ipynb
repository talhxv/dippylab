```python
import cv2 as cv #computer vision lib
import numpy as np #numpy to create the matrix 
import matplotlib.pyplot as plt #for plotting the images 

img_bgr = cv.imread('images.jpg')
#changing the image to rgb for pyplot 
img_rgb = cv.cvtColor(img_bgr, cv.COLOR_BGR2RGB)

#grayscaling the image for binarization
img_gray = cv.cvtColor(img_rgb, cv.COLOR_RGB2GRAY)
height, width = img_gray.shape

#creating a matrix of size h x w to binarize the image 
binarization = np.zeros((height, width), dtype=np.uint8)

#using for loop to iterate through the rows first then columns and then it checks whether the pixel value is less than 128 (127 inclusive) and if it is it makes it 0 and if it is higher than that it makes it 255, binarizing it in the process
for i in range(height):
    for j in range(width):
        if img_gray[i, j] <  128:
            binarization[i, j] = 0
        else:
            binarization[i, j] = 255
            
#using subplots to display the images together on one plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
ax1.axis('off')     
ax1.imshow(img_rgb)
ax1.set_title('Original Image')
ax2.axis('off')
ax2.imshow(binarization, cmap='gray')
ax2.set_title('Binarized Image')
plt.show()
```
