```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def dilate(image, kernel):
    return cv2.dilate(image, kernel, iterations=1)

def erode(image, kernel):
    return cv2.erode(image, kernel, iterations=1)

def close(image, kernel):
    dilated_image = dilate(image, kernel)
    closed_image = erode(dilated_image, kernel)
    return closed_image

def myMorphology(image_path):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    _, binary_image = cv2.threshold(image, 14, 255, cv2.THRESH_BINARY)
    kernel = np.ones((3, 3), np.uint8)
    closed_image = close(binary_image, kernel)
    eroded_image = erode(closed_image, kernel)
    boundaries = cv2.subtract(closed_image, eroded_image)
    return image, boundaries

# Example usage
original_image, boundaries = myMorphology('image.png')


plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(original_image, cmap='gray')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.title('Boundaries')
plt.imshow(boundaries, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np

img = cv2.imread('image2.png')

kernel = np.ones((11,3), np.uint8)
vertical = cv2.erode(img, kernel, iterations=1)

kernel2 = np.ones((3,11), np.uint8)
horizontal = cv2.erode(img, kernel2, iterations=2)

plt.subplots(1, 3, figsize=(20, 20))

plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(vertical)
plt.title('Vertical Erosion')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(horizontal)
plt.title('Horizontal Erosion')
plt.axis('off')

plt.show()
```
